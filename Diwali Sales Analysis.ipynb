```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading Data into Datframe

```python
df = pd.read_csv("D:\\DOWNLOADS\\Diwali Sales Data.csv", encoding="unicode_escape")
```

```python
df
```

# Data Cleaning

```python
df.info()
```

```python
df.drop(["Status", "unnamed1"], axis=1, inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

### Changing & Renaming datatype 

```python
df["Amount"] = df["Amount"].astype(int)
```

```python
df.rename(columns = {"Marital_Status": "Shaadi"}, inplace=True)
```

```python
df["Shaadi"].replace(to_replace='No', value="Unmarried", inplace=True)
df["Shaadi"].replace(to_replace="Yes", value="Married", inplace=True)
```

### Describe() 

```python
df.describe()
```

```python
df[["Age", "Orders", "Amount"]].describe()
```

# Exploratory Data Analysis

## ● Gender

#### -> Count the number of Male and Female

```python
sns.set(style="darkgrid")
ax = sns.countplot(x='Gender', data = df)

# to show numbers on the bar
for bar in ax.containers:
    ax.bar_label(bar)

plt.show()
```

#### -> Show sales by Gender

```python
sales_gen = df.groupby(["Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
sales_gen
```

```python
ax = sns.barplot(x="Gender", y="Amount", data = sales_gen)
plt.show()
```

### From the above analysis we can see that most of the buyers are Female and purchasing power of female is also greater then Males.

## ● Age Group

```python
df.columns
```

#### -> Which age group is more active 

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
ax1 = sns.countplot(x="Age Group", data=df, hue="Gender")
plt.title("Count of Age Group by Gender")
for bar in ax1.containers:
    ax1.bar_label(bar)

plt.subplot(1,2,2)
ax2 = sns.countplot(x="Age Group", data=df)
plt.title("Count of Age Group")
for bar in ax2.containers:
    ax2.bar_label(bar)
```

```python
plt.figure(figsize=(12,5))
ax1 = sns.countplot(x="Age Group", data=df, hue="Shaadi")
plt.title("Count of Age Group by Marital Status")
for bar in ax1.containers:
    ax1.bar_label(bar)
```

#### -> Which age group is sepnding more

```python
Sales_Age = df.groupby(["Age Group"], as_index=False)["Amount"].sum()
Sales_Age
```

```python
ax = sns.barplot(x="Age Group", y="Amount", data=Sales_Age)
```

### From the above graph we can see that Females of age group(26-35) are more active and more in purchasing.

## ● State

#### -> Total no. of orders from top 10 state.

```python
df.columns
```

```python
State_orders = df.groupby(["State"], as_index=False)["Orders"].sum().sort_values("Orders", ascending=False).head(10)
State_orders
```

```python
plt.figure(figsize=(16,5))
ax = sns.barplot(x="State", y="Orders", data=State_orders)
for bar in ax.containers:
    ax.bar_label(bar)
plt.title("Top 10 State by orders")
plt.show()
```

#### -> Total sales from top 10 states.

```python
State_sales = df.groupby(["State"], as_index=False)["Amount"].sum().sort_values("Amount", ascending=False).head(10)
State_sales
```

```python
plt.figure(figsize=(16,5))
sns.barplot(x="State", y="Amount", data=State_sales, palette="hsv_r")
plt.title("Top 10 State by Sales")
plt.show()
```

### From the above graph we can see that most of the orders are from Uttar Pradesh, Maharashtra and Karnataka.

## ● Marital Status

```python
df.columns
```

#### -> Count of Married and Unmarried

```python
ax = sns.countplot(x="Shaadi", data=df)
for bar in ax.containers:
    ax.bar_label(bar)
plt.title("Total Married and Unmarried")
plt.show()
```

#### -> Buyers according to marital status

```python
Marital_status = df.groupby(["Shaadi"], as_index=False)["Orders"].sum().sort_values("Orders", ascending=False)
Marital_status
```

```python
ax = sns.barplot(x="Shaadi", y="Orders", data=Marital_status, palette="cool")
for bar in ax.containers:
    ax.bar_label(bar)
plt.title("Orders by Married & Unmarried")
plt.show()
```

#### -> Buyers according to marital status grouped by Gender

```python
Marital_status_gender = df.groupby(["Shaadi","Gender"], as_index=False)["Orders"].sum().sort_values("Orders", ascending=False)
Marital_status_gender
```

```python
ax = sns.barplot(x="Shaadi", y="Orders", hue="Gender", data=Marital_status_gender, palette="Accent")
for bar in ax.containers:
    ax.bar_label(bar)
plt.title("Orders by Married & Unmarried grouped by Gender")
plt.show()
```

### From the above graph we can see that both Married & Unmarried Females have more orders then Males

## ● Occupation

```python
df.columns
```

#### -> Count of buyers from different occupation

```python
plt.figure(figsize=(12,5))
ax = sns.countplot(x="Occupation", data=df)
for bar in ax.containers:
    ax.bar_label(bar)

plt.xticks(rotation=45)
plt.show()
```

#### -> amount spend from different Occupation

```python
Sales_occupation = df.groupby(["Occupation"], as_index=False)["Amount"].sum().sort_values('Amount', ascending=True)
Sales_occupation
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x="Occupation", y="Amount", data=Sales_occupation)
plt.xticks(rotation=45)
plt.title("Total amount spend by buyers from different occupation")
plt.show()
```

### From above graph we can see that most of the buyers are from IT sector.
### and the amount spend are as follow IT Sector, Healthcare, Aviation...

## ● Product Category

```python
df.columns
```

#### -> Top 5 Product category by purchase order

```python
Top_category_orders = df.groupby(["Product_Category"], as_index=False)["Orders"].sum().sort_values("Orders", ascending=False).head(5)
Top_category_orders
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(x="Product_Category", y="Orders", data=Top_category_orders)
for bar in ax.containers:
    ax.bar_label(bar)
plt.title("Top 5 Product category by purchase order")
plt.show()
```

#### -> Top 5 product category by Sales

```python
Top_category_state = df.groupby(["Product_Category"], as_index=False)["Amount"].sum().sort_values("Amount", ascending=False).head(5)
Top_category_state
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(x="Product_Category", y="Amount", data=Top_category_state)
plt.show()
```

## From the above graph we can see that most orders are from "Clothing & Apparel" but more sales is from "Food"

## ● Customer

#### -> Top 10 Customers by sales

```python
Top_cust_sales = df.groupby(["Cust_name"], as_index=False)["Amount"].sum().sort_values("Amount", ascending=False).head(10)
Top_cust_sales
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(x="Cust_name", y="Amount", data = Top_cust_sales, palette="hsv")
for bar in ax.containers:
    ax.bar_label(bar)
plt.title("Top 10 Customers by sales", fontsize=15)
plt.show()
```

#### -> Top 10 Customers by orders

```python
Top_cust_orders = df.groupby(["Cust_name"], as_index=False)["Orders"].sum().sort_values("Orders", ascending=False).head(10)
Top_cust_orders
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(x="Cust_name", y="Orders", data = Top_cust_orders, palette="cool")
for bar in ax.containers:
    ax.bar_label(bar)
plt.title("Top 10 Customers by Orders", fontsize=15)
plt.show()
```

## From the above graph we can see that highest orders & sales are from "Vishakha" but 2nd highest orders are from "Shreyshi" and 2nd highest sales is from "Sudevi".

## ● Product

#### -> Top 10 most ordered product

```python
Top_product_orders = df.groupby(["Product_ID"], as_index=False)["Orders"].sum().sort_values("Orders", ascending=False).head(10)
Top_product_orders
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(x="Product_ID", y="Orders", data = Top_product_orders)
for bar in ax.containers:
    ax.bar_label(bar)
plt.title("Top 10 Most Ordered Products", fontsize=15)
plt.xticks(rotation=45)
plt.show()
```

#### -> Top 10 Highest Sales Product

```python
Top_product_sales = df.groupby(["Product_ID"], as_index=False)["Amount"].sum().sort_values("Amount", ascending=False).head(10)
Top_product_sales
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(x="Product_ID", y="Amount", data = Top_product_sales, palette="cool")
for bar in ax.containers:
    ax.bar_label(bar)
plt.title("Top 10 Products by Sales", fontsize=15)
plt.xticks(rotation=45)
plt.show()
```
